# 模型结果对比

```python
import time
from copy import copy
import pandas as pd
import os
import numpy as np
import cv2
from PIL import Image
import matplotlib
from matplotlib import pyplot as plt
plt.rcParams['lines.color'] = '#226383'
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d import proj3d
plt.rcParams['font.sans-serif']='SimHei'
plt.rcParams['axes.unicode_minus'] = False
```

```python
y1=[85.42,81.37,71.91,51.88,31.35] #$MAC_s(AWA2)$
y2=[44.69,68.36,62.51,40.3,8.18] #$MAC_t(AWA2)$
y3=[50.29,51.22,43.98,29.02,8.54] #$MAC_t(CUB)$
y4=[29.02,26.77,23.4,19.61,7.87]#$MAC_t(CUB)$
x=range(5)
label=[1e-4,2e-4,5e-4,10e-4,15e-4]
fig, ax1 = plt.subplots(1, 1,figsize=(10,7))    # a figure with a 2x1 grid of Axes
ax1.plot(x,y1,'r-s',lw=1)
ax1.plot(x,y2,'r--*',lw=1)
ax1.plot(x,y3,'g-s',lw=1)
ax1.plot(x,y4,'g--*',lw=1)
ax1.set_xticks([])     #去除坐标轴刻度
ax1.set_xticks(x)  #设置坐标轴刻度
ax1.set_xticklabels(labels=label,rotation=0,fontsize='small')  #设置刻度的显示文本，rotation旋转角度，fontsize字体大小

ax1.set_xlabel("Learning rate",fontsize=20)
ax1.set_ylabel("MCA(%)",fontsize=20)
ax1.set_title("Performance of varying learning rate ",fontsize=20)
ax1.legend(['$MAC_s(AWA2)$','$MAC_t(AWA2)$','$MAC_s(CUB)$','$MAC_t(CUB)$'],loc='upper right')
plt.show()
```

```python
y1=[38.3,36.1] #$MAC_s(AWA2)$
y2=[46.3,34.2] #VGG
y3=[31.26,21.2] #ResNet
y4=[39.49,26.73] #GoogLeNet
labels=['$MAC_s(CUB)$','$MAC_t(CUB)$']
x=[0,1]
total_width, n = 2, 12
width = total_width / n

fig, ax1 = plt.subplots(1, 1,figsize=(10,7))    # a figure with a 2x1 grid of Axes
rect=ax1.bar(x,y1,color="#ffd966",width=width,edgecolor="black")
for i,r in enumerate(rect):
    h=r.get_height()
    plt.text(r.get_x()+0.03,h,y1[i])
    
for i in range(len(x)):
    x[i] = x[i] + width+0.02
rect=ax1.bar(x,y2,color="#a9d18e",width=width,edgecolor="black")
for i,r in enumerate(rect):
    h=r.get_height()
    plt.text(r.get_x()+0.04,h,y2[i])
        
for i in range(len(x)):
    x[i] = x[i] + width+0.02
rect=ax1.bar(x,y3,color="#9dc3e6",width=width,edgecolor="black")
for i,r in enumerate(rect):
    h=r.get_height()
    plt.text(r.get_x()+0.07,h,y3[i])
    
for i in range(len(x)):
    x[i] = x[i] + width+0.02
rect=ax1.bar(x,y4,color="#f4b183",width=width,edgecolor="black")
for i,r in enumerate(rect):
    h=r.get_height()
    plt.text(r.get_x()+0.07,h,y4[i])

ax1.set_xticks([])     #去除坐标轴刻度
ax1.set_xticks((0.28,1.28))  #设置坐标轴刻度
ax1.set_xticklabels(labels=labels,rotation=0,fontsize='small')  #设置刻度的显示文本，rotation旋转角度，fontsize字体大小

# ax1.set_xticks(labels)
ax1.set_ylabel("MCA(%)",fontsize=20)
ax1.set_title("Performance of different subnets",fontsize=20)
ax1.legend(['$BUC\_MAC(MobileNetV2)$','$BUC\_MAC(VGG)$','$BUC\_MAC(ResNet)$','$BUC\_MAC(GoogLeNet)$'])
plt.show()
```
